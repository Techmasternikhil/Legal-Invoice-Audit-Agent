**🚀 LexAudit AI — Multi-Agent Invoice Compliance & Risk Intelligence System
Enterprise Agents Track**

LexAudit AI is my attempt to reimagine how companies audit invoices: a process that’s traditionally slow, inconsistent, and frustrating, even for experienced finance teams. Instead of manual checks, spreadsheets, and subjective judgment calls, LexAudit uses a coordinated set of agents to verify financial data, enforce policy rules, detect fraud patterns, and score vendor risk with the precision of an enterprise auditor.

This writeup explains the full system end-to-end and the thinking behind it, so a reader even without prior finance background can clearly understand what the agent is doing, why it’s built this way, and how each part contributes to the final automated decision.

**🔍 1. The Problem**

Every large business faces the same bottleneck: invoices pile up! Someone needs to validate dates, ensure GSTIN or tax IDs are correct, check descriptions, compare charges to policy rules, calculate risk, and write a compliance summary.

That “someone” is usually a human auditor and the task is almost always tedious.

A single invoice can take 30–45 minutes to properly review. Multiply that by hundreds per month, and you have a recipe for burnout, inconsistency, and unnecessary financial leakage.

**The issues show up in three ways:**
- slow processing
- inconsistent decision-making
- undetected overbilling or fraud.

Most companies only ever catch the obvious mistakes. Subtle violations are almost impossible to track manually.

I wanted to see if a multi-agent system could handle this entire process with more discipline, more consistency, and far more speed.

**✨ 2. The Idea: Fully Automated, Explainable Invoice Auditing**

LexAudit AI is built around a simple but ambitious question:

**"What if a team of specialized agents could do the entire audit pipeline — extraction, compliance, risk scoring, reasoning, and reporting without a human touching the document?"**

To make that possible, LexAudit performs the same steps a human analyst would, but with no fatigue and no variability in judgment. It reads the invoice, structures the content, evaluates it against your master policy, applies a risk model, and then writes a polished executive summary that looks like it came from a senior financial controller.

The entire process takes under 10 seconds.



**🧠 3. Why Agents Are the Perfect Fit**

The invoice domain is messy. Documents vary by vendor, region, format, and clarity. A single rule-engine script can’t understand vague descriptions or contextual cues.

* Agents solve this because they can reason through ambiguity.

* In LexAudit, each agent is deliberately designed to think about a single responsibility:

* One agent focuses purely on understanding the document.

* Another is trained to analyze policy violations.

* Another calculates risk using vendor reputation and invoice signals.

* One writes the final executive report.

Each step feeds into the next. The entire system feels like watching a real team collaborate.

LexAudit’s workflow moves through five major stages:

**1. Intake Agent**

Extracts clean, raw text from the uploaded PDF to prepare it for structured analysis.

**2. Classification Agent**

Validates if the document is a real invoice and extracts vendor, dates, currency, and line items into JSON.

**3. Compliance Agent**

Audits every line item against financial policy rules to identify violations and non-compliant charges.

**4. Risk Engine**

Calculates a final 0–100 risk score using vendor history, violation patterns, and high-value red flags.

**5. Reporting Agent**

Generates a polished, auditor-style executive summary with findings, savings, and a final recommendation.

```python
%%writefile app.py
import streamlit as st
import os
import json
import time
import re
import google.generativeai as genai
import pandas as pd
from datetime import datetime

# --- PAGE CONFIGURATION ---
st.set_page_config(
    page_title="LexAudit AI Enterprise",
    page_icon="⚖️",
    layout="wide",
    initial_sidebar_state="expanded"
)

# --- THEME CONFIGURATION ---
if 'theme' not in st.session_state:
    st.session_state.theme = 'dark'

def toggle_theme():
    if st.session_state.theme == 'light':
        st.session_state.theme = 'dark'
    else:
        st.session_state.theme = 'light'

# Define color palettes
themes = {
    "light": {
        "bg_color": "#ffffff",
        "secondary_bg_color": "#f8fafc",
        "text_color": "#0f172a", # Sharp dark blue-black for readability
        "metric_bg": "#ffffff",
        "metric_border": "#e2e8f0", 
        "metric_shadow": "rgba(148, 163, 184, 0.1)",
        "sidebar_bg": "#f1f5f9", # Slightly darker gray for better separation
        "input_bg": "#ffffff",
        "input_border": "#cbd5e1",
        "uploader_bg": "#f8fafc", 
        "button_text": "#0f172a",
        "icon_color": "#0f172a" 
    },
    "dark": {
        "bg_color": "#0e1117",
        "secondary_bg_color": "#1e293b",
        "text_color": "#fafafa",
        "metric_bg": "#1e293b",
        "metric_border": "#334155",
        "metric_shadow": "rgba(0, 0, 0, 0.3)",
        "sidebar_bg": "#262730",
        "input_bg": "#262730",
        "input_border": "#4b5563",
        "uploader_bg": "#1e293b",
        "button_text": "#fafafa",
        "icon_color": "#fafafa" 
    }
}

current_theme = themes[st.session_state.theme]

# --- STYLING ---
st.markdown(f"""
<style>
    /* --- Global & Layout --- */
    [data-testid="stAppViewContainer"] {{
        background-color: {current_theme['bg_color']};
        color: {current_theme['text_color']};
    }}
    
    [data-testid="stSidebar"] {{
        background-color: {current_theme['sidebar_bg']};
        border-right: 1px solid {current_theme['metric_border']};
    }}
    
    [data-testid="stHeader"] {{
        background-color: rgba(0,0,0,0);
    }}

    /* --- Sidebar Toggle Arrow Fix --- */
    [data-testid="stSidebarCollapsedControl"] svg, 
    [data-testid="stSidebarExpandedControl"] svg {{
        fill: {current_theme['icon_color']} !important;
        color: {current_theme['icon_color']} !important;
    }}
    
    /* Fix specifically for the SVG path inside the sidebar button */
    [data-testid="stSidebarCollapsedControl"] svg path, 
    [data-testid="stSidebarExpandedControl"] svg path {{
        fill: {current_theme['icon_color']} !important;
    }}

    /* --- Text Elements --- */
    h1, h2, h3, p, li, span, div, button, input, textarea, label {{
        font-family: 'Inter', sans-serif;
        color: {current_theme['text_color']};
    }}
    
    .material-icons, .icon {{ font-family: 'Material Icons' !important; }}
    
    .main-header {{ 
        font-size: 2.2rem; 
        font-weight: 700; 
        margin-bottom: 0;
        background: -webkit-linear-gradient(45deg, #2563eb, #db2777);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
    }}
    
    .sub-header {{ 
        font-size: 1rem; 
        opacity: 0.8;
        margin-bottom: 2rem; 
    }}
    
    /* --- Cards & Metrics --- */
    .metric-card {{ 
        background-color: {current_theme['metric_bg']}; 
        padding: 1rem; 
        border-radius: 12px; 
        border: 1px solid {current_theme['metric_border']}; 
        box-shadow: 0 4px 6px -1px {current_theme['metric_shadow']}; 
        text-align: center; 
        transition: transform 0.2s ease;
    }}
    
    .metric-card:hover {{
        transform: translateY(-2px);
        box-shadow: 0 10px 15px -3px {current_theme['metric_shadow']};
    }}
    
    .metric-value {{ 
        font-size: 1.5rem; 
        font-weight: 700; 
        color: #3b82f6 !important; 
        white-space: nowrap; 
        overflow: hidden;
        text-overflow: ellipsis;
    }}
    
    .metric-label {{ 
        font-size: 0.75rem; 
        font-weight: 600; 
        opacity: 0.7; 
        text-transform: uppercase;
        letter-spacing: 0.5px;
        margin-top: 0.5rem;
    }}
    
    /* --- Input Fields --- */
    .stTextInput input, .stTextArea textarea {{
        background-color: {current_theme['input_bg']} !important;
        color: {current_theme['text_color']} !important;
        border: 1px solid {current_theme['input_border']} !important;
        border-radius: 8px;
    }}
    
    /* --- File Uploader Styling --- */
    /* 1. Dropzone container */
    [data-testid='stFileUploader'] section {{
        background-color: {current_theme['uploader_bg']} !important;
        border: 2px dashed {current_theme['input_border']} !important;
        border-radius: 12px;
    }}
    
    /* 2. Text color inside uploader */
    [data-testid='stFileUploader'] section span, 
    [data-testid='stFileUploader'] section div, 
    [data-testid='stFileUploader'] section small,
    [data-testid='stFileUploader'] section p {{
        color: {current_theme['text_color']} !important;
    }}

    /* 3. Cloud Icon (Only target the SVG inside the dropzone section) */
    [data-testid='stFileUploader'] section svg {{
        fill: {current_theme['text_color']} !important;
    }}
    
    /* 4. Browse Button */
    [data-testid='stFileUploader'] section button {{
        background-color: {current_theme['metric_bg']} !important;
        color: {current_theme['button_text']} !important;
        border: 1px solid {current_theme['input_border']} !important;
    }}

    /* 5. Uploaded File Item List - Ensure text visible */
    [data-testid='stFileUploader'] ul {{
        background-color: transparent !important;
    }}
    
    /* 6. File Item Text - Force color */
    [data-testid='stFileUploader'] ul li div {{
        color: {current_theme['text_color']} !important;
    }}
    
    /* 7. Delete 'X' Icon - Let it inherit color (Fixes white square issue) */
    [data-testid='stFileUploader'] ul li svg {{
        fill: {current_theme['text_color']} !important;
    }}

    /* --- Standard Buttons --- */
    .stButton > button {{
        background-color: {current_theme['metric_bg']} !important;
        color: {current_theme['button_text']} !important;
        border: 1px solid {current_theme['input_border']} !important;
        border-radius: 8px;
        transition: all 0.3s ease;
        font-weight: 600;
    }}
    
    .stButton > button:hover {{
        border-color: #3b82f6 !important;
        color: #3b82f6 !important;
        background-color: {current_theme['input_bg']} !important;
    }}
    
    /* Primary Button Override */
    .stButton > button[kind="primary"] {{
        background-color: #3b82f6 !important;
        color: white !important;
        border: none !important;
        box-shadow: 0 4px 6px -1px rgba(59, 130, 246, 0.3);
    }}
    
    .stButton > button[kind="primary"]:hover {{
        box-shadow: 0 6px 8px -1px rgba(59, 130, 246, 0.4);
    }}

    /* --- Risk Badges --- */
    .risk-badge-high {{ background-color: #fee2e2; color: #991b1b; padding: 4px 8px; border-radius: 4px; font-weight: 600; font-size: 0.8rem; }}
    .risk-badge-med {{ background-color: #fef3c7; color: #92400e; padding: 4px 8px; border-radius: 4px; font-weight: 600; font-size: 0.8rem; }}
    .risk-badge-low {{ background-color: #dcfce7; color: #166534; padding: 4px 8px; border-radius: 4px; font-weight: 600; font-size: 0.8rem; }}
    
    /* Report Box Styling */
    .report-box {{ background-color: {current_theme['metric_bg']}; padding: 1.5rem; border-radius: 12px; border: 1px solid {current_theme['metric_border']}; margin-top: 20px; }}
</style>
""", unsafe_allow_html=True)

# --- HELPER FUNCTIONS ---
def clean_amount(val):
    if isinstance(val, (int, float)): return val
    if isinstance(val, str):
        clean_str = re.sub(r'[^\d.-]', '', val)
        try: return float(clean_str)
        except ValueError: return 0
    return 0

def calculate_risk_score(audited_data, vendor_name):
    # Simple mock vendor DB
    VENDOR_DB = {
        "Flipkart India Private Limited": {"RiskProfile": "Low", "HistoryScore": 95},
        "CANVAS WEAVES": {"RiskProfile": "Medium", "HistoryScore": 70},
        "Unknown Vendor": {"RiskProfile": "High", "HistoryScore": 40}
    }
    
    base_score = 100
    vendor_info = VENDOR_DB.get(vendor_name, VENDOR_DB["Unknown Vendor"])
    vendor_risk_deduction = (100 - vendor_info["HistoryScore"]) * 0.25
    
    violation_count = sum(1 for item in audited_data if item.get('FlaggedAmount', 0) > 0)
    violation_deduction = min(50, violation_count * 10) 
    
    high_value_deduction = 0
    for item in audited_data:
        if clean_amount(item.get('Total', 0)) > 10000:
            high_value_deduction = 15
            break
            
    final_score = base_score - vendor_risk_deduction - violation_deduction - high_value_deduction
    return max(0, round(final_score))

def get_risk_level(score):
    if score >= 80: return "Low", "risk-badge-low"
    if score >= 50: return "Medium", "risk-badge-med"
    return "High", "risk-badge-high"

def generate_warning_explanation(results):
    """
    Generates a natural language explanation for why the Manual Review/Auto-Reject warning is shown.
    """
    reasons = []
    
    # 1. Risk Score Reason
    if 50 <= results['risk_score'] <= 80:
        reasons.append(f"The automated risk score ({results['risk_score']}/100) falls within the manual review threshold (50-80).")
    
    # 2. Violation Count Reason
    violation_count = sum(1 for item in results['audited_items'] if item.get('FlaggedAmount', 0) > 0)
    if violation_count > 0:
        reasons.append(f"There are {violation_count} specific line items flagged for potential policy violations.")
        
    # 3. High Value Reason
    for item in results['audited_items']:
        if clean_amount(item.get('Total', 0)) > 10000:
            reasons.append("High-value item detected (>10,000).")
            break
            
    if not reasons:
        return "The invoice patterns generated a medium confidence score requiring standard quality assurance check."
        
    return " ".join(reasons)

# --- AI AGENTS ---
@st.cache_resource
def get_models(api_key):
    genai.configure(api_key=api_key)
    text_model = genai.GenerativeModel(
        model_name="gemini-2.5-flash-preview-09-2025",
        generation_config={"temperature": 0.2, "response_mime_type": "application/json"}
    )
    # Additional model for report generation (text mode)
    report_model = genai.GenerativeModel(
        model_name="gemini-2.5-flash-preview-09-2025",
        generation_config={"temperature": 0.7} # Higher temp for better writing
    )
    return text_model, report_model

def run_intake_agent(uploaded_file):
    if not uploaded_file: return None
    try:
        import pypdf
        pdf_reader = pypdf.PdfReader(uploaded_file)
        text = ""
        for page in pdf_reader.pages: text += page.extract_text() + "\n"
        return text
    except Exception as e:
        st.error(f"PDF Error: {e}")
        return None

def run_classification_agent(model, text):
    # UPDATED PROMPT: Adds "IsInvoice" check
    prompt = f"""
    Act as a Forensic Invoice Auditor. Analyze this document text.
    
    STEP 1: DOCUMENT CLASSIFICATION
    Is this a valid commercial Invoice, Bill, or Receipt?
    - If YES, set "IsInvoice": true
    - If NO (e.g. it is a Resume, Contract, Letter, Report, or random text), set "IsInvoice": false and stop extraction.
    - If UNSURE, default to true.
    
    STEP 2: EXTRACTION (Only if IsInvoice is true)
    Extract line items. Detect Vendor Name and Currency Symbol.
    
    Return JSON structure: 
    {{ 
      "IsInvoice": true/false,
      "Vendor": "...", 
      "Date": "...", 
      "Items": [
         {{ "Date": "...", "Description": "...", "Total": 0.00, "Currency": "$", "Category": "..." }}
      ]
    }}
    
    Document Text:
    \"\"\"{text}\"\"\"
    """
    try:
        response = model.generate_content(prompt)
        return json.loads(response.text)
    except: return {"IsInvoice": True} # Default to True on error

def run_compliance_agent(model, invoice_data, guidelines):
    current_date = datetime.now().strftime("%Y-%m-%d")
    prompt = f"""
    Act as a Chief Compliance Officer. Audit these items against the guidelines.
    Context Date: {current_date}
    
    Guidelines: \"\"\"{guidelines}\"\"\"
    Vendor: {invoice_data.get('Vendor', 'Unknown')}
    Items: {json.dumps(invoice_data.get('Items', []))}
    
    INSTRUCTIONS:
    For EACH item, check if it violates a rule.
    If Violated: Set 'Status'='Rejected', 'FlaggedAmount'=deduction, 'Reason'=explanation, 'RuleRef'=rule number.
    If Approved: Set 'Status'='Approved', 'FlaggedAmount'=0, 'Reason'='Compliant', 'RuleRef'='N/A'.
    
    Return JSON array of audited items.
    """
    try:
        response = model.generate_content(prompt)
        data = json.loads(response.text)
        for item in data:
            item['FlaggedAmount'] = clean_amount(item.get('FlaggedAmount', 0))
        return data
    except: return []

def run_output_agent(model, vendor_name, risk_score, total_savings, currency, audited_items):
    """
    Generates a professional Executive Summary Report in Markdown.
    """
    prompt = f"""
    You are a Senior Financial Auditor. Write a concise Executive Audit Report for this invoice.
    
    Vendor: {vendor_name}
    Risk Score: {risk_score}/100
    Total Savings Identified: {currency}{total_savings}
    
    Audit Findings (JSON):
    {json.dumps(audited_items)}
    
    OUTPUT FORMAT (Markdown):
    ### 📋 Executive Summary
    [Write a professional paragraph summarizing the invoice compliance status, vendor risk, and key findings.]
    
    ### 🚩 Key Violations & Adjustments
    [List the top 3 most significant violations found, if any. Use bullet points. Be specific about why it was flagged.]
    
    ### 💡 Auditor Recommendation
    [Final recommendation: Approve, Reject, or Pay Partial. Provide a clear next step for the finance team.]
    """
    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"⚠️ Error generating report: {str(e)}" 

# --- UI LAYOUT ---
with st.sidebar:
    st.header("⚙️ System Settings")
    
    btn_label = "🌙" if st.session_state.theme == 'light' else "☀️"
    if st.button(btn_label):
        toggle_theme()
        st.rerun()

    api_key_input = st.text_input("Gemini API Key", type="password", value=os.environ.get("GEMINI_API_KEY", ""))
    
    st.subheader("📜 Master Policy")
    guidelines = st.text_area("Rules Engine", height=400, value="""1. VENDOR CONTEXT:
- Law Firms: Admin/Overhead is NON-BILLABLE.
- Retail/E-commerce: Handling/Platform fees are APPROVED.
- Education: Exam/Tuition fees are APPROVED.

2. TRAVEL POLICY:
- Goods Transport (GT) approved at actuals.
- Professional Services Travel Time capped at 50% of hourly rate.

3. DOCUMENTATION:
- "Misc Charges" without description > 500 is REJECTED.
- Tax Invoice must show GSTIN.
- Future Dated Invoices are REJECTED.

4. TIME RESTRICTIONS:
- Dinner expenses before 8:00 PM are Personal (REJECT).""")

st.markdown('<p class="main-header">⚖️ LexAudit </p>', unsafe_allow_html=True)
st.markdown('<p class="sub-header">AI-Powered Compliance & Risk Engine</p>', unsafe_allow_html=True)

uploaded_file = st.file_uploader("Upload Invoice (PDF)", type=['pdf'])

if 'audit_results' not in st.session_state:
    st.session_state.audit_results = None

if st.button("🚀 Analyze Invoice", type="primary", disabled=not uploaded_file):
    if not api_key_input:
        st.error("API Key missing")
        st.stop()
        
    text_model, report_model = get_models(api_key_input)
    
    with st.spinner("🤖 AI Agents working..."):
        # 1. Intake
        raw_text = run_intake_agent(uploaded_file)
        
        # 2. Classify
        invoice_structure = run_classification_agent(text_model, raw_text)
        
        # Rate Limiting Fix: 15s pause before next heavy call
        time.sleep(15) 

        if not invoice_structure.get("IsInvoice", False):
            st.error("⛔ REJECTED: Document is NOT a valid Invoice.")
            st.session_state.audit_results = None 
            st.stop()

        vendor_name = invoice_structure.get('Vendor', 'Unknown')
        
        # 3. Compliance
        audited_items = run_compliance_agent(text_model, invoice_structure, guidelines)
        
        # Rate Limiting Fix: 15s pause
        time.sleep(15)

        # 4. Risk Engine
        risk_score = calculate_risk_score(audited_items, vendor_name)
        risk_label, risk_css = get_risk_level(risk_score)
        
        # Calc Totals & Safe Currency Extraction
        total_billed = sum(clean_amount(i.get('Total', 0)) for i in audited_items)
        total_savings = sum(i.get('FlaggedAmount', 0) for i in audited_items)
        
        items = invoice_structure.get('Items', [])
        currency = items[0].get('Currency', '₹') if items else '₹'
        
        # Generate Text Report
        executive_report = run_output_agent(report_model, vendor_name, risk_score, total_savings, currency, audited_items)

        st.session_state.audit_results = {
            "risk_score": risk_score,
            "risk_label": risk_label,
            "risk_css": risk_css,
            "vendor_name": vendor_name,
            "currency": currency,
            "total_billed": total_billed,
            "total_savings": total_savings,
            "audited_items": audited_items,
            "executive_report": executive_report
        }

# --- DISPLAY RESULTS ---
if st.session_state.audit_results:
    results = st.session_state.audit_results
    
    decision_color = "#16a34a" if results['risk_score'] > 80 else "#d97706" if results['risk_score'] > 50 else "#dc2626"
    decision_text = "✅ APPROVED" if results['risk_score'] > 80 else "⚠️ MANUAL REVIEW" if results['risk_score'] > 50 else "❌ AUTO-REJECT"
    if results['total_savings'] > 0 and results['risk_score'] > 50: decision_text = "⚠️ RE-CHECK THE INVOICE"

    # Optimised Background Logic for Light/Dark Modes
    if st.session_state.theme == "light":
        if results['risk_score'] > 80: bg_color = "#dcfce7"
        elif results['risk_score'] > 50: bg_color = "#fef3c7"
        else: bg_color = "#fee2e2"
    else:
        bg_color = f"{decision_color}15"

    st.markdown(f"""
    <div style="padding: 20px; border-radius: 10px; border: 1px solid {decision_color}; background-color: {bg_color}; margin-bottom: 20px;">
        <h2 style="color: {decision_color}; margin:0;">{decision_text}</h2>
        <p style="opacity: 0.8; color: {current_theme['text_color']};">Confidence Score: 92% • Vendor: <b>{results['vendor_name']}</b></p>
    </div>
    """, unsafe_allow_html=True)
    
    # --- EXPLANATION SECTION FOR WARNINGS/REJECTIONS ---
    if "APPROVED" not in decision_text or results['risk_score'] <= 80:
        explanation = generate_warning_explanation(results)
        st.info(f"**Why this decision?**\n\n{explanation}", icon="ℹ️")
    # ---------------------------------------------------

    c1, c2, c3, c4 = st.columns(4)
    c1.markdown(f"""<div class="metric-card"><div class="metric-value">{results['risk_score']}/100</div><div class="metric-label"><span class="{results['risk_css']}">{results['risk_label']} Risk</span></div></div>""", unsafe_allow_html=True)
    c2.markdown(f"""<div class="metric-card"><div class="metric-value">{results['currency']}{results['total_billed']:,.0f}</div><div class="metric-label">Invoice Total</div></div>""", unsafe_allow_html=True)
    c3.markdown(f"""<div class="metric-card"><div class="metric-value" style="color:#dc2626 !important">{results['currency']}{results['total_savings']:,.0f}</div><div class="metric-label">Rejected Amount</div></div>""", unsafe_allow_html=True)
    c4.markdown(f"""<div class="metric-card"><div class="metric-value">{len(results['audited_items'])}</div><div class="metric-label">Lines Audited</div></div>""", unsafe_allow_html=True)

    # --- EXECUTIVE REPORT SECTION ---
    st.markdown(f'<div class="report-box">', unsafe_allow_html=True)
    st.markdown(results['executive_report'])
    st.markdown('</div>', unsafe_allow_html=True)
    # --------------------------------

    st.subheader("🔍 Detailed Audit Findings")
    
    df = pd.DataFrame(results['audited_items'])
    if not df.empty:
        # --- FIX: Ensure columns exist before selection ---
        required_cols = ['Reason', 'RuleRef', 'Description', 'Total', 'Status', 'FlaggedAmount']
        for col in required_cols:
            if col not in df.columns:
                df[col] = "N/A" # Fill missing columns to prevent KeyError
        # --------------------------------------------------

        display_df = df[['Description', 'Total', 'Status', 'FlaggedAmount', 'Reason', 'RuleRef']].copy()
        st.dataframe(display_df, use_container_width=True)
        
        csv = df.to_csv(index=False).encode('utf-8')
        st.download_button("📥 Download Detailed Report (CSV)", csv, f"audit_report_{results['vendor_name'].replace(' ', '_')}.csv", "text/csv")
    
    with st.expander("👨‍⚖️ Auditor Override (Adaptive Learning)"):
        col_f1, col_f2 = st.columns([3, 1])
        with col_f1: feedback_item = st.selectbox("Select disputed item", df['Description'].tolist() if not df.empty else [])
        with col_f2: 
            if st.button("Mark as Correct"): st.success("Feedback recorded.")
```
